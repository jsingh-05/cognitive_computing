```python
#A1
import numpy as np
import matplotlib.pyplot as plt

M=int(input("Enter a value: "))
x=np.linspace(-10,10)
y1=M*x*x
y2=M*np.sin(x)
plt.title("Mathematical Functions")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y1,'r--',label='y = M*x^2')
plt.plot(x,y2,'b-',label='y = M*sin(x)')
```

```python
#A2
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data={'Subject':['Maths','English','Physics','Chemistry','Biology'],
      'Marks':[85,93,95,96,82]}
df=pd.DataFrame(data)
colours=['#FF6347','#32CD32','#4682B4','#8A2BE2','#FFD700']
plt.grid()
sns.barplot(x='Subject',y='Marks',data=df,palette=colours)
plt.title("Marks of Students in Different Subjects")
for i in range(df.shape[0]):
    plt.text(i, df['Marks'][i] + 1, str(df['Marks'][i]), ha='center')
```

```python
#A3
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(102317177)
data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10,8))

axs[0, 0].plot(np.cumsum(data))
axs[0, 0].set_title('Cumulative Sum Line Plot')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Sum')
axs[0, 0].grid(True)

axs[0, 1].scatter(range(50), data)
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Random Values')
axs[0, 1].grid(True)

axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
#A4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/refs/heads/main/company_sales_data.csv'
df = pd.read_csv(url)

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o', color='green')
plt.title('Company Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    sns.lineplot(data=df, x='month_number', y=product, label=product)
plt.title('Monthly Sales Data for Each Product')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.show()

average_values = df.mean(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.barplot(x=average_values.index, y=average_values.values, palette='Set2')
plt.title('Average Monthly Values of Each Feature')
plt.xlabel('Features')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
