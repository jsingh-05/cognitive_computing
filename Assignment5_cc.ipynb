```python
#A3(a)(b)(c)(d)
import numpy as np

initials = "JS"
X = sum(ord(char) for char in initials)
sales = np.array([X, X+50, X+100, X+150, X+200])

tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)

discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

weeks = 3
sales_matrix = np.tile(sales, (weeks, 1))
increase_factors = np.array([1.00, 1.02, 1.04]).reshape(weeks, 1)
adjusted_sales_matrix = sales_matrix * increase_factors

print("Sales Data:", sales)
print("Taxed Sales:", taxed_sales)
print("Discounted Sales:", discounted_sales)
print("Expanded Sales Data:", adjusted_sales_matrix)
```
