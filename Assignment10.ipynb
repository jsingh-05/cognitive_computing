```python
#A1
import re
import nltk
from nltk.corpus import stopwords
from nltk.probability import FreqDist
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

text = "Reading is a fundamental skill that opens doors to new worlds. Whether it's fiction or non-fiction, books allow us to explore, learn, and grow. In a digital age, the value of reading has only increased, as it offers an escape and fosters creativity."
text_cleaned = re.sub(r'[^\w\s]', '', text.lower())

words = nltk.word_tokenize(text_cleaned)
sentences = nltk.sent_tokenize(text)

words_split = text_cleaned.split()
words_tokenized = nltk.word_tokenize(text_cleaned)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_tokenized if word not in stop_words]

freq_dist = FreqDist(filtered_words)

print(text_cleaned)
print(words)
print(sentences)
print(words_split)
print(words_tokenized)
print(filtered_words)
print(freq_dist)
```

```python
#A3
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

reviews = [
    "This product is amazing! It works perfectly.",
    "Not great. It broke within a week of use.",
    "The product is okay, but I expected more features."
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews)

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(reviews)

print(vectorizer.get_feature_names_out())
print(tfidf_vectorizer.get_feature_names_out())
print(tfidf_matrix)
for i, row in enumerate(tfidf_matrix.toarray()):
    top_indices = np.argsort(row)[-3:][::-1]
    top_keywords = [tfidf_vectorizer.get_feature_names_out()[index] for index in top_indices]
    print(f"Review {i+1} - Top Keywords: {top_keywords}")
```

```python
#A4
from sklearn.metrics.pairwise import cosine_similarity
texts = [
    "AI is revolutionizing industries by automating tasks and providing intelligent insights.",
    "Blockchain technology ensures secure transactions and transparency in various applications."
]

tokenized_texts = [nltk.word_tokenize(text.lower()) for text in texts]

set1 = set(tokenized_texts[0])
set2 = set(tokenized_texts[1])
jaccard_similarity = len(set1.intersection(set2)) / len(set1.union(set2))

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])

print(tokenized_texts)
print(f"Jaccard Similarity: {jaccard_similarity}")
print(f"Cosine Similarity: {cosine_sim[0][0]}")
```

```python
#A5
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

reviews = [
    "This product is amazing! It works perfectly.",
    "Not great. It broke within a week of use.",
    "The product is okay, but I expected more features."
]

for review in reviews:
    blob = TextBlob(review)
    print(f"Review: {review}")
    print(f"Polarity: {blob.sentiment.polarity}, Subjectivity: {blob.sentiment.subjectivity}")

for review in reviews:
    blob = TextBlob(review)
    if blob.sentiment.polarity > 0:
        print(f"Review: {review} -> Positive")
    elif blob.sentiment.polarity < 0:
        print(f"Review: {review} -> Negative")
    else:
        print(f"Review: {review} -> Neutral")

positive_reviews = [review for review in reviews if TextBlob(review).sentiment.polarity > 0]
positive_text = ' '.join(positive_reviews)

wordcloud = WordCloud().generate(positive_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#A6
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import random

text = ("Reading is a powerful habit that opens doors to endless knowledge and imagination. "
        "It allows people to explore new ideas, cultures, and experiences from the comfort of their homes. "
        "Through books and stories, readers gain insight into the lives of others and learn about the world around them. "
        "Reading strengthens the mind, improves focus, and expands vocabulary in meaningful ways. "
        "Whether for education or entertainment, reading remains a timeless activity that enriches lives and inspires growth.")


tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

input_sequences = []
words = text.lower().split()

for i in range(1, len(words)):
    n_gram_sequence = words[:i+1]
    token_list = tokenizer.texts_to_sequences([" ".join(n_gram_sequence)])[0]
    input_sequences.append(token_list)

max_sequence_len = max([len(seq) for seq in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

X = input_sequences[:, :-1]
y = tf.keras.utils.to_categorical(input_sequences[:, -1], num_classes=total_words)

model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_sequence_len - 1))
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=0)

def generate_text(seed_text, next_words=3):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')

        predicted_probs = model.predict(token_list, verbose=0)
        predicted = np.argmax(predicted_probs, axis=-1)[0]

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += " " + word
                break
    return seed_text

seed_word = "reading"
generated = generate_text(seed_word, next_words=4)
print("Generated Text:\n", generated)
```
