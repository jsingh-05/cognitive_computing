```python
#A1
import string
import nltk
from nltk.tokenize import PunktSentenceTokenizer, WordPunctTokenizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

nltk.download('stopwords')

sentence_tokenizer = PunktSentenceTokenizer(train_text="Painting is great")
word_tokenizer = WordPunctTokenizer()

paragraph = """Painting is a powerful form of expression that allows artists to communicate emotions, stories, and ideas without the need for words. Each brushstroke captures a moment of thought, a burst of inspiration, or a quiet reflection. Through color, texture, and composition, painting can transport viewers to imagined worlds or offer new perspectives on familiar scenes. It bridges time and culture, preserving both personal and collective histories. Whether abstract or realistic, every painting carries the unique fingerprint of its creator, making each piece a one-of-a-kind dialogue between artist and audience."""

lower_text = paragraph.lower()
clean_text = lower_text.translate(str.maketrans('', '', string.punctuation))

sentence_tokens = sentence_tokenizer.tokenize(paragraph)
word_tokens = word_tokenizer.tokenize(clean_text)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokens if word not in stop_words]

fdist = FreqDist(filtered_words)

print("\nOriginal Text:\n", paragraph)
print("\nCleaned Text:\n", clean_text)
print("\nSentence Tokens:\n", sentence_tokens)
print("\nWord Tokens:\n", word_tokens)
print("\nFiltered Words (No Stopwords):\n", filtered_words)
print("\nWord Frequency:\n")
fdist.plot(10, title="Word Frequency Distribution (Excluding Stopwords)")
```

```python
#A2
import nltk
import string
from nltk.tokenize import PunktSentenceTokenizer, WordPunctTokenizer
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords, wordnet
import matplotlib.pyplot as plt
nltk.download('wordnet')

word_tokenizer = WordPunctTokenizer()
sentence_tokenizer = PunktSentenceTokenizer(train_text="Painting is great.")
paragraph = """Painting is a powerful form of expression that allows artists to communicate emotions, stories, and ideas without the need for words. Each brushstroke captures a moment of thought, a burst of inspiration, or a quiet reflection. Through color, texture, and composition, painting can transport viewers to imagined worlds or offer new perspectives on familiar scenes. It bridges time and culture, preserving both personal and collective histories. Whether abstract or realistic, every painting carries the unique fingerprint of its creator, making each piece a one-of-a-kind dialogue between artist and audience."""

lower_text = paragraph.lower()
clean_text = lower_text.translate(str.maketrans('', '', string.punctuation))
word_tokens = word_tokenizer.tokenize(clean_text)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokens if word not in stop_words]

porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

print("Word, PorterStemmer, LancasterStemmer, Lemmatizer")
print("-" * 70)
for word in filtered_words:
    porter_stem = porter.stem(word)
    lancaster_stem = lancaster.stem(word)
    lemma = lemmatizer.lemmatize(word)
    print(word + ", " + porter_stem + ", " + lancaster_stem + ", " + lemma)
```

```python
#A3
import re

paragraph = """Painting is a powerful form of expression that allows artists to communicate emotions, stories, and ideas without the need for words. Each brushstroke captures a moment of thought, a burst of inspiration, or a quiet reflection. Through color, texture, and composition, painting can transport viewers to imagined worlds or offer new perspectives on familiar scenes. It bridges time and culture, preserving both personal and collective histories. Whether abstract or realistic, every painting carries the unique fingerprint of its creator, making each piece a one-of-a-kind dialogue between artist and audience."""

words_gt5 = re.findall(r'\b\w{6,}\b', paragraph)

numbers = re.findall(r'\b\d+\b', paragraph)

capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', paragraph)

alpha_words = re.findall(r'\b[a-zA-Z]+\b', paragraph)

vowel_words = re.findall(r'\b[aeiouAEIOU][a-zA-Z]*\b', paragraph)

print("Words with more than 5 letters:\n", words_gt5)
print("\nNumbers in the text:\n", numbers)
print("\nCapitalized words:\n", capitalized_words)
print("\nWords containing only alphabets:\n", alpha_words)
print("\nWords starting with a vowel:\n", vowel_words)
```

```python
#A4
import re

text = """Painting is a powerful form of expression that allows artists to communicate emotions, stories, and ideas without the need for words. Each brushstroke captures a moment of thought, a burst of inspiration, or a quiet reflection. Through color, texture, and composition, painting can transport viewers to imagined worlds or offer new perspectives on familiar scenes. It bridges time and culture, preserving both personal and collective histories. Whether abstract or realistic, every painting carries the unique fingerprint of its creator, making each piece a one-of-a-kind dialogue between artist and audience.
Contact me at example@mail.com or visit https://www.example.com. Call 123-456-7890 or +91 9876543210 for more info."""

def clean_and_replace(text):
    text = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', '<EMAIL>', text)
    text = re.sub(r'https?://\S+|www\.\S+', '<URL>', text)
    text = re.sub(r'(\+?\d{1,3}[\s-]?)?\d{3}[-\s]?\d{3}[-\s]?\d{4}', '<PHONE>', text)
    return text

def custom_tokenizer(text):
    pattern = r"""
        \b\w+(?:-\w+)+\b
        | \b\w+'\w+\b
        | \b\d+\.\d+\b
        | \b\w+\b
    """
    tokens = re.findall(pattern, text, flags=re.VERBOSE)
    return tokens

clean_text = clean_and_replace(text)
tokens = custom_tokenizer(clean_text)

print("Cleaned Text:\n", clean_text)
print("\nTokens:\n", tokens)
```
